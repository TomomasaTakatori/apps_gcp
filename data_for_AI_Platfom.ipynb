# Auto ML table用にデータを変える

```
new_train_data = pd.concat([pd.Series(name='type'), train_data], axis=1)
new_train_data.head()
```

```
new_train_data['type'] = "TRAIN"
new_train_data.set_index('type', inplace=True)
new_train_data.head()
```

```
# 面倒なので typeを追加する関数を書く
def set_type_as_index(df, typ):
    new_df = pd.concat([pd.Series(name='type'), df], axis=1)
    new_df['type'] = typ
    new_df.set_index('type', inplace=True)
    return new_df
```

```
new_train_data = set_type_as_index(train_data, "TRAIN")
new_valid_data = set_type_as_index(validation_data, "VALIDATE")
new_test_data = set_type_as_index(test_data, "TEST")
```
